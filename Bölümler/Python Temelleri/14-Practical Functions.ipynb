```python
def divideNumber(number):
    return number/2
```

```python
divideNumber(20)
```

```python
myList = [3,5,7,10,20,30]
```

```python
myResultList = []
for element in myList:
    myResultList.append(divideNumber(element))
```

```python
myResultList
```
