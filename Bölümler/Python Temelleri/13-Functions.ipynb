```python
#args, kwargs (arguments, key word arguments)
```

```python
def args_sum(*args): # to be able to send as much input as you want.
    return sum(args)
```

```python
args_sum(10)
```

```python
args_sum(10,20)
```

```python
args_sum(10,20,30,40)
```

```python
def args_example(*args):
    print(args)
```

```python
args_example(10,20,30,40)
```

```python
def kwargs_example(**kwargs):
    print(kwargs)
```

```python
kwargs_example(apple=100,banana=150,melon=200)
```

```python
def kwargs_example2(**kwargs):
    if "apple" in kwargs:
        print("applooooo")
    else:
        print(":(")
        
```

```python
kwargs_example2(apple=10,banana=15)
```

```python
kwargs_example2(banana=111)
```
