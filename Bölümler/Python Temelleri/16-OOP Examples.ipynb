```python
class Musician():
    def __init__(self,name):
        self.name = name
        print("musician class")
    def test1(self):
        print("tes1")
    def test2(self):
        print("test2")
```

```python
atil = Musician("Atil")
```

```python
atil.name
```

```python
atil.test1()
```

```python
atil.test2()
```

```python
class MusicianPlus(Musician): # to be a child of Musician class we can use parent class name within the child class.
   # thanks to the same init we will not write attributes
    def __init__(self,name):
        Musician.__init__(self,name)
        print("musician plus")
    def test3(self):
        print("test3")
    def test1(self): # this is knwon as overriding
        print ("new test1")
```

```python
atlas = MusicianPlus("Atlas")
```

```python
atlas.test1()
```

```python
atlas.test2()
```

```python
atlas.name = "Atlas Samancioglu"
```

```python
type(atlas)
```

```python
atlas.test3()
```

```python
class Banana():
    def __init__(self,name):
        self.name = name
    def info (self):
        return f"100 calories {self.name}"
class Apple():
    def __init__(self,name):
        self.name = name
    def info (self):
        return f"150 calories {self.name}"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.info()
```

```python
apple.info()
```

```python
fruitList = [banana,apple]
```

```python
for fruit in fruitList:
    print(fruit.info())
```

```python
#to imprison, to block access to the variable
```

```python
class Phone():
    def __init__(self, name, price):
        self.name = name
        self.__price = price   # private attribute. Accessible only within the same class.

    def info(self):
        print(f"{self.name} price is:{self.__price}")

    def changePrice(self, price):
        self.__price = price


```

```python
iphone = Phone("iphone 11",500)
```

```python
iphone.info()
```

```python
iphone.__price = 300 # we could not change it 
```

```python
iphone.info()
```

```python
iphone.changePrice(700)
```

```python
iphone.info()
```

```python
 # It is like a blueprint.
```

```python
from abc import ABC, abstractmethod
```

```python
class Car(ABC):
    @abstractmethod
    def maxSpeed():
        pass
        
```

```python
class Tesla(Car):
    pass
    
```

```python
#  tesla = Tesla()  we have to use methods that are within the abstract method 
```

```python
class Tesla(Car):
    def maxSpeed(self):
        print("200 km")
    
```

```python
tesla = Tesla()
```

```python
tesla.maxSpeed()
```

```python
class Mercedes(Car):
    def maxSpeed(self):
        print("150 km ")
```

```python
# thanks to the abstract classes we can not forget any method that are going to be used
```

```python
# Special methods
```

```python
class Fruit():
    def __init__(self,name,calories):
        self.name = name
        self.calories = calories
    def __str__(self):
        return f"{self.name}:{self.calories} calories" 
    def __len__(self):
        return self.calories
        
```

```python
myFruit = Fruit("banana",150)
```

```python
myFruit.calories
```

```python
myFruit.name
```

```python
print(myFruit)
```

```python
# by using __str__ we can print properties of object
print(myFruit)
```

```python
class Train():
    def __init__(self,name):
        self.name = name
    def __getitem__(self,key):
        return self.name
```
