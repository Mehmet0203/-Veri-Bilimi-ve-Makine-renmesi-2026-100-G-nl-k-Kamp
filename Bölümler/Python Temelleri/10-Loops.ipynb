```python
my_tuple = (10,20,30,40,50,60,70)
```

```python
my_list = list((10,20,30,40,50,60,70))
```

```python
for num in my_list:
    print(num)
```

```python
40 in my_list
```

```python
# to search and find then stop program for efficiency.
```

```python
for number in my_list:
    if number == 40: 
        continue # when condition is true it will continue with for and 40 will not be writed.
    print(number)
```

```python
for number in my_list:
    if number == 40: 
        print(number)
    continue 
```

```python
for number in my_list:
    if number == 40: 
        break # to finish the loop
    print(number)
```

```python
for num in my_list:
    pass # pass is used to fill or arrange the code later.
```
