```python
class Person(): # Generally class name consists of capital letter.
    #properties
    name = ""
    age = 0
    gender = ""
    #initializer method
    #we can use different parameter names or we can use with self keyword.
    def __init__(self,name,age,gender): #Constructor. we can use  our attributes as parameters
        self.name = name # self name indicates class's name
        self.age = age
        self.gender = gender
        #method
    def test(self):
        print("test")
```

```python
#atil = Person() we have to put parameters otherwise we will get erors
```

```python
atil = Person("Atil",35,"Male")
```

```python
type(atil)
```

```python
atil.name
```

```python
class Person(): # Generally class name consists of capital letter.
    #properties
  #  name = ""
    #age = 0
  #  gender = ""
    job = "developer" # we can write arbitary properties here. And we can give default values.
# in python we don't have to write properties it can understand itself.
    
    #initializer method
    def __init__(self,name,age,gender): # we can use here for obligatory properties
        self.name = name
        self.age = age
        self.gender = gender
        
        #method
    def printName(self):
        print(self.name)
```
