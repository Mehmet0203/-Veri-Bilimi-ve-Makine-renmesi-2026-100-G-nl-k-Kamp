```python
def process_value(value: int | str) -> str:
    if isinstance(value, int):
        return f"processed integer: {value}"
    else:
        return f"processed string: {value}"
```

```python
process_value(50)
```

```python
process_value("ASF")
```

```python
process_value([10,20,30])
```
