```python
while True:
    try:
        myAge = int(input("Enter your age : "))
        print(myAge * 2)
        #break
    
    except:
        print("please enter your age !")
    else:
        print("else executed")
        break
    finally:
        print("Finally  ") # if we want to execute anything always we can use this keyword
```
