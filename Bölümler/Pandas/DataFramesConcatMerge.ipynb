```python
# merge
```

```python
df_merge1 = pd.read_csv("7-merge_data1.csv")
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv("7-merge_data2.csv")
```

```python
df_merge2
```

```python
# merge - outer join
```

```python
df_merged_outer = pd.merge(df_merge1,df_merge2,on = "Employee_ID", how = "outer") # to merge with common column
```

```python
df_merged_outer # for not found values nan is used.
```

```python
# df_merged_left
```

```python
df_merged_left = pd.merge(df_merge1,df_merge2, on ="Employee_ID", how = "left") # it only take cares of first table 
```

```python
df_merged_left 
```

```python
 df_merged_right = pd.merge(df_merge1,df_merge2, on ="Employee_ID", how = "right") # based on second table 
```

```python
df_merged_right
```

```python
# merge inner
```

```python
merge_inner = pd.merge(df_merge1,df_merge2, on ="Employee_ID",how ="inner") # it takes intersection set from both table 
```

```python
merge_inner
```
