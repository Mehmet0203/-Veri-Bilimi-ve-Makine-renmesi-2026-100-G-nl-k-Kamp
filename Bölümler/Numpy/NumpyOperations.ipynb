```python
#z-score
```

```python
data = np.array([10,12,13,15,18,25,100,105])# 100 and 105 are outlier
```

```python
#outlier
```

```python
mean = np.mean(data)
```

```python
mean 
```

```python
data.sum()
```

```python
std = np.std(data)
```

```python
std
```

```python
z_scores= (data - mean) / std
```

```python
z_scores
```

```python
data[z_scores > 1]
```
