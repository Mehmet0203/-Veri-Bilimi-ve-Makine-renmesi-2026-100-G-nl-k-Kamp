```python
plt.scatter("Height","Weight", data = data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete Weight vs Height")
plt.show()
```

```python
sns.set_style("whitegrid") # with grids and we can use without it too.
sns.scatterplot(x = "Height",y = "Weight",data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
# this one is better than upper one as you can see
```

```python
sns.set_style("darkgrid") # with grids and we can use without it too.
sns.scatterplot(x = "Height",y = "Weight",data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("white") 
sns.scatterplot(x = "Height",y = "Weight", hue = "Sex", data = data)# used to emphasize specif thing.
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("white") 
sns.scatterplot(x = "Height",y = "Weight", hue = "Medal", data = data)# used to emphasize specif thing.
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("white")

sns.scatterplot(
    x="Height",
    y="Weight",
    hue="Sex",
    style="Medal",
    data=data
)

plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")

plt.show()
```

```python
sns.set_style("white")

sns.scatterplot(
    x="Height",
    y="Weight",
    hue="Sex",
    style="Medal",
    size = "Age", 
    data=data
)

plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")

plt.show()
```

```python
sns.set_style("whitegrid")
sns.lineplot(x = "Height", y = "Weight", hue = "Sex", data = data)
plt.xlabel("Height of Athletes")
plt.ylabel("Weight of Athletes")
plt.title("Athletes Weight vs Height")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x = "Height", hue = "Sex", data = data)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x = "Height", hue = "Sex", kind = "kde" , data = data) # kde used visualize distrubution of observation
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
plt.show()
```

```python
sns.barplot(x = "Medal", y = "Height", hue = "Sex", data = data )
plt.title("Medals by Height")
plt.show()
```
