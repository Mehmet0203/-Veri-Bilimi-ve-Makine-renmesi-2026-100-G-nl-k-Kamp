```python
numpy_arr1 = np.linspace(0,10,20) # from 0 to 10 it returns 20 elements
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2)
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b*")
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b*-")
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b--")
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2,"b-*")
plt.show()
```

```python
plt.subplot(1, 2, 1)   # 1 row, 2 columns, first graph
plt.plot(numpy_arr1, numpy_arr2, "b-*")
plt.subplot(1, 2, 2)   # second graph
plt.plot(numpy_arr2, numpy_arr1, "r--")
plt.show()
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.5,0.5])  # first two indicates spaces , last two indicates height and width
figure_axes.plot(numpy_arr1,numpy_arr2,"g")
figure_axes.set_xlabel("X Axis ")
figure_axes.set_ylabel("Y Axis ")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure(dpi = 100) # dpi increases its resolution

new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])

new_axes.plot(numpy_arr1, numpy_arr1**2, label="numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1**3, label="numpy array ** 3")
new_axes.legend(loc=0) # we can indicate its location within the paranthesis
plt.show()
```
